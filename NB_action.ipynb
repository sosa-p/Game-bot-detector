```python
import pandas as pd
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
import sklearn.metrics as mt
from sklearn.model_selection import cross_val_score, cross_validate
```

```python
#탐지할 데이터 셋 설정
action_data =pd.read_csv('Player actions features.csv')
data = action_data[['collect_max_count','Sit_ratio','Abyss_get_ratio','Money_get_ratio','Exp_get_ratio','Item_get_ratio']]
Type1 = []

for each in action_data['Type']:
    if each == 'Human':
        Type1.append(1)
    else:
        Type1.append(0)
        
action_data['Type'] = Type1
target = action_data['Type']
data1 = action_data['Use_portal_count']

data.head()
```

```python
#훈련셋, 테스트셋 분리
X_train, X_test, Y_train, Y_test = train_test_split(data, target, random_state=0)
```

```python
#Classifier 선언
clf = GaussianNB()
clf.fit(X_train, Y_train)
pred = clf.predict(X_test) 
```

```python
#정확도 출력
scores = cross_val_score(clf, X_train, Y_train, cv=5)
print('cross validation avg: ', scores.mean())   

accuracy = mt.accuracy_score(Y_test, pred)
recall = mt.recall_score(Y_test, pred)
precision = mt.precision_score(Y_test, pred)
f1_score = mt.f1_score(Y_test, pred)

print('Accuracy: ', format(accuracy))
print('Recall: ', format(recall))
print('Precision: ', format(precision))
print('F1_score: ', format(f1_score))
```

```python
#탐지 결과 출력
print("------ Result------")
i = 1
human = 0
bot = 0
for each in pred :
    if each == 1:
        print("NO.%d is Human." %(i))
        human+=1
    else:
        print("NO.%d is Bot." %(i))
        bot+=1
    i+=1
print("Human : %d" %human)
print("Bot : %d" %bot)
print("total : %d" %i)
```
